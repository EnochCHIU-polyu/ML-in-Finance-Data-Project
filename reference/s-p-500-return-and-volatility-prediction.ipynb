This notebook aims to predict the daily returns and volatility of the S&P 500 index using regression and variance modeling.

```python
import yfinance as yf
import pandas as pd

# Download S&P 500 data with adjusted close
data = yf.download('^GSPC', start='2010-01-01', end='2025-01-01', auto_adjust=True)

# Flatten columns if necessary
if isinstance(data.columns, pd.MultiIndex):
    data.columns = ['_'.join(col).strip() for col in data.columns.values]

# Display first 5 rows
data.head()
```

```python
# Compute daily return
data['Return'] = data['Close_^GSPC'].pct_change()

# Compute rolling features
data['MA10'] = data['Close_^GSPC'].rolling(window=10).mean()
data['STD10'] = data['Return'].rolling(window=10).std()

# Drop missing values
data = data.dropna()

# Display first few rows
data.head()
```
